```python
from pathlib import Path
from amc_parser.motion import MotionCapture
from amc_parser import parse_asf, parse_amc
import copy

asf_path = Path.cwd() / 'mocap' / 'subjects' / '02' / '02.asf'
amc_path = Path.cwd() / 'mocap' / 'subjects' / '02' / '02_02.amc'

joints = parse_asf(asf_path)
frames = parse_amc(amc_path)
```

```python
# Build a dictionary with the same format as frames[0], but all values are 0
# Use this to plot the human skeleton in all 0s pose
zero_pose = {joint: [0] * len(frames[0][joint]) for joint in frames[0].keys()}
```

```python
joints['root'].set_motion(zero_pose)
joints["root"].draw()
```

```python
# Now we can experiment with changing the pose of the skeleton

turned = copy.deepcopy(zero_pose)
turned['lowerback'][0] = 45


joints['root'].set_motion(turned)
joints["root"].draw()
```
